```python
import pandas as pd

# Load the CSV file
file_path = '/home/chiara/Energy/Data/Ausgrid Load Data.csv'  # Replace with the path to your CSV file
save_path = '/home/chiara/Energy/Data/Transl.csv'  # Replace with the path to save the new CSV file

data = pd.read_csv(file_path, header=0, sep=';')
# Transpose the data
transposed_data = data.T

data = transposed_data.iloc[:, 0:1440]  
data.to_csv(save_path, index=True, header=True, sep=';')
```

```python
x_columns = [f'x_{j+1}' for j in range(72)]
y_columns = [f'y_{j+1}' for j in range(72)]
Columns = ['index'] + x_columns + y_columns
df= pd.DataFrame(columns=Columns)
for index in data.index:
    print(index)
    row_idx = data.index.get_loc(index)  # Get the integer position of the row
    #k=0
    for i in range(0,data.shape[1], 72):
        #print(k)
        #k=k+1
        if i + 72 > 1439:
            break
        #print(f'x_i : {i} i_stop : {i + 71}')
        #print(f'y_i : {72 + i} i_stop : {72 + i + 72}')
        x_values = data.iloc[row_idx, i:i + 72].values
        y_values = data.iloc[row_idx, (72 + i):(72 + i + 72)].values
        
        row_data = [f'{index}_{int(i/72)}'] + list(x_values) + list(y_values)
        
        
        df.loc[len(df)] = row_data
```

```python

save_path = '/home/chiara/Energy/Data/Train_2m_72.csv'  # Replace with the path to save the new CSV file

df.to_csv(save_path, index=True, header=True, sep=';')
```

```python
import pandas as pd

# Load the CSV file
file_path = '/home/chiara/Energy/Data/Ausgrid Load Data.csv'  # Replace with the path to your CSV file
save_path = '/home/chiara/Energy/Data/TEST.csv'  # Replace with the path to save the new CSV file

data = pd.read_csv(file_path, header=0, sep=';')
# Transpose the data
transposed_data = data.T

data = transposed_data.iloc[:, 1478:(1478+240+72)]
```

```python
x_columns = [f'x_{j+1}' for j in range(72)]
y_columns = [f'y_{j+1}' for j in range(72)]
Columns = ['index'] + x_columns + y_columns
df= pd.DataFrame(columns=Columns)
for index in data.index:
    print(index)
    row_idx = data.index.get_loc(index)  # Get the integer position of the row
    #k=0
    for i in range(0,217, 72):
        if i + 72 > (72*3):
            break
        #print(k)
        #k=k+1
        #print(f'x_i : {i} i_stop : {i + 71}')
        #print(f'y_i : {72 + i} i_stop : {72 + i + 72}')
        x_values = data.iloc[row_idx, i:i + 72].values
        y_values = data.iloc[row_idx, (72 + i):(72 + i + 72)].values
        
        # Ensure the length of row_data matches the number of columns in df
        if len(x_values) == 72 and len(y_values) == 72:
            row_data = [f'{index}_{int(i/72)}'] + list(x_values) + list(y_values)
            df.loc[len(df)] = row_data
        else:
            print(f"Skipping row {index} due to mismatched data lengths: x_values={len(x_values)}, y_values={len(y_values)}")


save_path = '/home/chiara/Energy/Data/Test_2m_72.csv'  # Replace with the path to save the new CSV file

df.to_csv(save_path, index=True, header=True, sep=';')
```
